## Parse json data into columns

```python
dictionary = df.event_json.apply(json.loads)
dictionary_df = pd.json_normalize(dictionary)
dictionary_df.isnull().sum()
```

Информация о доходе присутствует всегда ->
Процент доходов содержащий информацию о городе = 62% [(860079-323708)/860079]

```python
df_parsed = pd.concat([df, dictionary_df], axis=1)
# delete not relevant columns
df_parsed = df_parsed.drop(['event_json','event_name', 'country_iso_code'], axis=1)
df_parsed.to_csv('df_parsed.csv')
```
